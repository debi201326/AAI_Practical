Practical 7: Implement simple programs to control the robot's movements and demonstrate the basic functionalities of the hardware.

```python
import matplotlib.pyplot as plt
import math

# Initial position and direction
x, y = 0, 0
direction = 0  # Angle in degrees (0 = right)
# Store path for plotting
path_x = [x]
path_y = [y]

def move_forward(steps=1):
    global x, y, direction
    rad = math.radians(direction)
    for _ in range(steps):
        x += round(math.cos(rad), 2)
        y += round(math.sin(rad), 2)
        path_x.append(x)
        path_y.append(y)
    print(f"Moved forward to ({x}, {y})")

def move_backward(steps=1):
    global x, y, direction
    rad = math.radians(direction)
    for _ in range(steps):
        x -= round(math.cos(rad), 2)
        y -= round(math.sin(rad), 2)
        path_x.append(x)
        path_y.append(y)
    print(f"Moved backward to ({x}, {y})")

def turn_left(angle=90):
    global direction
    direction = (direction + angle) % 360
    print(f"Turned left. Now facing {direction} degrees")

def turn_right(angle=90):
    global direction
    direction = (direction - angle) % 360
    print(f"Turned right. Now facing {direction} degrees")

def stop():
    print("Robot stopped.")

# Run a simple program
move_forward(3)
turn_left()
move_forward(2)
turn_right()
move_backward(1)
stop()

# Plot path
plt.figure(figsize=(6,6))
plt.plot(path_x, path_y, marker='o')
for i, (x_pos, y_pos) in enumerate(zip(path_x, path_y)):
    plt.text(x_pos, y_pos, str(i), ha='center', va='bottom')
plt.title("Robot Movement Simulation")
plt.xlabel("X")
plt.xlabel("Y")
plt.grid(True)
plt.show()
```
